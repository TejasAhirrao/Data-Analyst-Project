```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10,12))
city_count=data['city'].value_counts().reset_index()
print(city_count)
city_count.columns=['city','city_count']
print(city_count)
sns.barplot(x='city_count',y='city',data=city_count)
plt.title('Matches play by city')
plt.show()
```
